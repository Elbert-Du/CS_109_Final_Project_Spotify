```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import json
```

```python
a = open("data/playlist_to_track_20171203.p",'rb')
tracks_in_playlists = pickle.load(a)

a = open("data/playlist_level_data_20171203.p", 'rb')
playlist_data = pickle.load(a)

a = open("data/song_level_data_audio_20171203.p", 'rb')
track_audio_data = pickle.load(a)

a = open("data/song_level_data_pop_20171203.p", 'rb')
track_other_data = pickle.load(a)
```

```python
track_audio_data.head()
```

```python
playlist_info = dict()
for playlist in tracks_in_playlists.items():
    all_audio_info = pd.DataFrame()
    all_track_info = pd.DataFrame()
    skip = 0
    #print(playlist)
    for track in playlist[1]:
        try:
            all_audio_info = all_audio_info.append(track_audio_data.loc[track,:])
            all_track_info = all_track_info.append(track_other_data.loc[track,:])
        except KeyError:
            print("track not found, abandoning playlist")
            skip = 1
            pass
    if skip == 1:
        continue
    agg_audio_info = all_audio_info.aggregate("std")
    agg_track_info = all_track_info.aggregate("std")
    try:
        playlist_info[playlist[0]] = pd.concat([playlist_data.loc[playlist[0],:], agg_audio_info, agg_track_info], sort = False)
    except KeyError:
        print("playlist not found")
        pass
```
